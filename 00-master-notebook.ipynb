```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
def read_raw_data(filename):
    data = pd.read_csv(filename)
    
    return data
```

```python
DATA_PATH = "data/"
```

```python
dataset = read_raw_data(DATA_PATH + "application_train.csv")
dataset
```

### Pick only necessary columns

```python
dataset.columns.tolist()
```

```python
PICK_COLS = ["SK_ID_CURR", "NAME_CONTRACT_TYPE", "CODE_GENDER", "FLAG_OWN_CAR",
            "AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE", "TARGET"]
```

```python
dataset = dataset[PICK_COLS].copy()
```

```python
dataset
```

```python
dataset["TARGET"].value_counts()
```

```python
dataset["NAME_CONTRACT_TYPE"].value_counts()
```

```python
len(dataset)
```

```python
DATA_DROP = len(dataset) - 2000
```

```python
arr_indices_top_drop = np.random.default_rng().choice(dataset.index, size=DATA_DROP, replace=False)
```

```python
sample_dataset = dataset.drop(index=arr_indices_top_drop)
```

```python
sample_dataset
```

```python
sample_dataset["TARGET"].value_counts()
```

```python
sample_dataset["NAME_CONTRACT_TYPE"].value_counts()
```

```python
joblib.dump(sample_dataset, "data/processed/sample_raw_dataset.pkl")
```
